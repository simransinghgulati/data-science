```python
import matplotlib.pyplot as plt
import numpy as np
```

### Themes

```python
plt.style.available
```

```python
plt.style.use("ggplot")
```

### Line Plot

```python
x = np.arange(-10,11)
line = 2*x + 20
curve = x**2
```

```python
plt.plot(x,line,color="red",label="Line",linestyle="dashed")
plt.plot(x,curve,color="green",label="Square",marker="o")

plt.title("Graph 1")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.show()
```

```python
theta = np.linspace(-2*np.pi,2*np.pi)

x=np.cos(theta)
y=np.sin(theta)
plt.plot(x,y)
plt.show()
```

### Scatter Plot

```python
plt.style.use("dark_background")

plt.scatter([1,2,2,2.5,1.5],[10,11,11.5,10,12],label="Scatter Points")

plt.title("Graph 2")
plt.xlim(0,5)
plt.ylim(5,15)
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.show()
```

### Bar Graph

```python
plt.style.use("seaborn")
plt.figure(figsize=(10,10))

plt.bar([0,2,4],[10,20,30],width=0.5, label="2018", tick_label=["Gold", "Platinum", "Silver"])
plt.bar([0.5,2.5,4.5],[10,15,50],width=0.5,label = "2019")

plt.xlabel("Element")
plt.ylabel("Tons")
plt.title("Graph 3")
plt.legend()

plt.show()
```

### Pie Chart

```python
plt.style.use("ggplot")

plt.figure(figsize=(5,5))

subjects = "Python", "Machine Learning", "Statistics", "R"
marks = [30,50,20,10]

plt.pie(marks,labels=subjects,explode=(0,0,0.05,0))

plt.title("Graph 4")
plt.show()
```

### Histogram

#### Standard Normal Distribution

```python
np.random.seed(42)
Xsn = np.random.randn(100)
sigma = 5
mean = 60

X1=np.round(Xsn*sigma + mean)
X2=np.round(Xsn*10 + 50)
```

```python
X1
```

```python
X2
```

#### Plotting N-Dist Data

```python
plt.style.use("classic")

plt.hist(X1, alpha=0.8, label="Statistics")
plt.hist(X2, alpha=0.7, label="Programming")

plt.xlim(20,80)
plt.ylim(0,40)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.title("Graph 5")

plt.show()
```
